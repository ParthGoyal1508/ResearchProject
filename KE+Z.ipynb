```
import pandas as pd 
from sklearn.utils import shuffle
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.decomposition import PCA
```

```
df = pd.read_csv('data_kz_combined.csv')
df = shuffle(df)
```

```
x = df.drop(['0'], axis=1)
y = df['0']
print(x.shape)
print(y.shape)
```

```
def calculate_accuracy(y_predict_model, y): 
  for i in range (len(y_predict_model)): 
    if y_predict_model[i] < 0.5: 
      y_predict_model[i] = 0
    else: 
      y_predict_model[i] = 1

  accurate = 0 
  total = len(y_predict_model) 
  for i in range(len(y_predict_model)): 
    if y_predict_model[i] == y[i]: 
      accurate += 1 
  return (accurate/total)
```

```
model = Sequential([
  Dense(330, activation='tanh', input_shape=(402,)),
  Dense(1, activation='sigmoid'),
])

# Compile the model.
model.compile(
  optimizer='adam',
  loss='binary_crossentropy',
  metrics=['accuracy'],
)

model.fit(
  x, # training data
  y, # training targets
  epochs=2,
  batch_size=56,
)

y_predict_model = model.predict(x)
accuracy_model1 = calculate_accuracy(y_predict_model, y)
print(accuracy_model1)
```

```
model2 = Sequential([
  Dense(330, activation='tanh', input_shape=(402,)),
  Dense(340, activation='tanh'),
  Dense(1, activation='sigmoid'),
])

# Compile the model.
model2.compile(
  optimizer='adam',
  loss='binary_crossentropy',
  metrics=['accuracy'],
)

model2.fit(
  x, # training data
  y, # training targets
  epochs=2,
  batch_size=56,
)

y_predict_model2 = model2.predict(x)
accuracy_model2 = calculate_accuracy(y_predict_model2, y)
print(accuracy_model2)
```
